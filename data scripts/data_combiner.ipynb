```python
import pandas as pd
import os
```

```python
directory = 'parsed_data'
input_files = [os.path.join(directory, o) for o in os.listdir(directory) if o[-4:] == '.csv']
input_files
```

```python
def memory_usage(df):
    return(round(df.memory_usage(deep=True).sum() / 1024 ** 2, 2))
```

```python
df_list = []

for file in input_files:
    print(file)
    try:
        df_list.append(pd.read_csv(file))
    except:
        continue
```

```python
df = pd.concat(df_list, sort=False, copy=False, ignore_index=True).sort_values(by=["name", "ddate"])
```
